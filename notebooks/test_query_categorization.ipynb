```python
def categorize_query(sql):
    sql_upper = sql.upper()

    if "DATE_TRUNC" in sql_upper or 'extract' in sql_upper:
        return "TIME_SERIES"
    elif "ORDER BY" in sql_upper and "LIMIT" in sql_upper:
        return "RANKING"
    elif "JOIN" in sql_upper:
        return "JOIN"
    elif "GROUP_BY" in sql_upper:
        return "GROUP_BY"
    elif "SUM(" in sql_upper or "AVG(" in sql_upper or "MAX(" in sql_upper or "MIN(" in sql_upper:
        return "AGGREGATION"
    elif "COUNT(*)" in sql_upper:
        return "COUNT"
    else:
        return "FILTER"
```

```python
print(categorize_query("select count(*), (timezone) from customers  limit 3"))
```
